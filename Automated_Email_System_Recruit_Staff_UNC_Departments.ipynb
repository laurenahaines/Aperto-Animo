```python
import pandas as pd
import smtplib

# set YOUR name 
# 'First Last'
your_name = ''

# set YOUR Gmail address
# 'address@gmail.com'
your_email = ''

# set YOUR major
# 'Public Policy'
your_major = ''

# set YOUR Gmail app password (not your normal password)
# Don't know what an app password is? Read this: https://support.google.com/accounts/answer/185833?hl=en
# 'abcdabcdabcdabcd'
your_password = ''

# set the path to the contact spreadsheet (.xlxs)
# 'C:\\Users\\yourname\\OneDrive - University of North Carolina at Chapel Hill\\Desktop\\Contacts.xlsx'
contacts = ''

# set the publication year
# 'YYYY-YYYY'
publication_year = ''

# set the submission deadline
# 'Month DD, YYYY'
deadline = ''

# set the link to the submission form
# 'https://linklinklink'
link = ''
```

```python
# We enter your Gmail account
server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
server.ehlo()
server.login(your_email, your_password)

# We read the <contacts> spreadsheet
email_list = pd.read_excel(contacts)
email_list = email_list.fillna('')
all_names = email_list['Name']
all_emails = email_list['Email']
all_departments = email_list['Department']
all_titles = email_list['Title']

# We send the recruitment email to each contact in <contacts>
for idx in range(len(all_emails)):

    name = all_names[idx]
    email = all_emails[idx]
    department = all_departments[idx]
    title = all_titles[idx]

    full_email = (
       
        "From: {0} <{1}>\n"
        "To: {3} <{4}>\n"
        "Subject: Academic Writing Opportunity for {6} Students\n\n"
        
        "Dear {6} {5},\n\n"
        
        "I hope this email finds you well."
        " I am a {2} major at UNC-Chapel Hill, writing to bring your attention to a great opportunity for students in the {6} department."
        " The annual, student-run undergraduate philosophy journal, Aperto Animo, is accepting staff applications for the upcoming publication year: ({7}).\n\n"
        
        "As the Editor-in-Chief of Aperto Animo, I believe it's an excellent opportunity for {6} students to get involved in a meaningful project and gain experience in academic writing."
        " Although Aperto Animo is expressly a philosophy journal, many of the submissions we recieve are multi-disciplinary."
        " Ergo, we strongly encourage students from any department to apply.\n\n"
        
        "I would greatly appreciate it if you could take a moment to share this information with students in your department and encourage them to consider applying for a staff position at Aperto Animo.\n\n"
        
        "For your convenience, here is a sample notification message:\n\n"
        
        "***\n\n"
        
        "The UNC-Chapel Hill student-run philosophy journal, Aperto Animo, is accepting applications for staff positions until {8}."
        " This is a great opportunity to get involved in a meaningful project and gain valuable experience in academic writing."
        " Although Aperto Animo is expressly a philosophy journal, many of the submissions it recieves are multi-disciplinary."
        " Ergo, students of any major are encouraged to apply."
        " Advanced knowledge of philosophy is not required.\n\n"
        
        "Apply here: {9} \n\n"
        
        "For more information, visit the Aperto Animo web page: https://philosophy.unc.edu/undergraduate/aperto-animo/ \n\n"
        
        "***\n\n"
        
        "Thank you for your consideration."
        " I hope this opportunity will be of interest to your students." 
        " Please let me know if you have any questions. \n\n"
        
        "Best regards,\n\n"
        
        "{0} \n"
                  
        # {0} your name, {1} your email, {2} your major
        # {3} their name, {4} their email, {5} their title, {6} their department
        # {7} publication year, {8} application deadline, {9} application link
        .format(your_name, your_email, your_major, name, email, title, department, publication_year, deadline, link))

    try:
        server.sendmail(your_email, [email], full_email)
        print('Email to {} successfully sent!\n\n'.format(email))
    except Exception as e:
        print('Email to {} could not be sent :( because {}\n\n'.format(email, str(e)))

# We exit your Gmail account
server.close()
```
