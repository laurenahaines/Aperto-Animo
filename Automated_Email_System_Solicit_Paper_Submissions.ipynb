```python
import pandas as pd
import smtplib

# set YOUR name 
# 'First Last'
your_name = ''

# set YOUR Gmail address
# 'address@gmail.com'
your_email = ''

# set YOUR Gmail app password (not your normal password)
# Don't know what an app password is? Read this: https://support.google.com/accounts/answer/185833?hl=en
# 'abcdabcdabcdabcd'
your_password = ''

# set the path to the contact spreadsheet (.xlxs)
# 'C:\\Users\\yourname\\OneDrive - University of North Carolina at Chapel Hill\\Desktop\\Contacts.xlsx'
contacts = ''

# set the publication year
# 'YYYY-YYYY'
publication_year = ''

# set the submission deadline
# 'Month DD, YYYY'
deadline = ''

# set the link to the submission form
# 'https://linklinklink'
link = ''
```

```python
''' Run this cell ONCE before changing the contact spreadsheet. 
Running multiple times will send duplicate emails to the contacts in your contact spreadsheet.'''

# connect to your Gmail account
server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
server.ehlo()
server.login(your_email, your_password)

# read contacts
email_list = pd.read_excel(contacts)
email_list = email_list.fillna('')
all_universities = email_list['University']
all_departments = email_list['Department']
all_titles = email_list['Title']
all_emails = email_list['Email']

# for each row in contacts ...
for idx in range(len(all_emails)):

    university = all_universities[idx]
    department = all_departments[idx]
    title = all_titles[idx]
    email = all_emails[idx]

    full_email = (
       
        "From: {0} <{1}>\n"
        "To: <{5}>\n"
        "Subject: Publication Opportunity for {3} Undergrads\n\n"
        
        "Dear {3} {4},\n\n"
        
        "I hope this email finds you well."
        " I'm contacting you on behalf of the undergraduate philosophy journal Aperto Animo (ISSN 2694-667X), affiliated with the Philosophy Dept. at the University of North Carolina, Chapel Hill."
        " We're excited to offer your students the opportunity to have their academic writing on philosophy, theology, political science, or related multidisciplinary studies published in our {6} issue.\n\n"
        
        "We DO NOT charge students a submission or publication fee."
        " Aperto Animo is a student-run, not-for-profit publication registered with the Library of Congress."
        " For more details, you can visit our UNC webpage:\n\n"
        
        "https://philosophy.unc.edu/undergraduate/aperto-animo/\n\n"
        
        "We would appreciate you forwarding this message to your students:\n\n"
        
        "***** SAMPLE MESSAGE BEGINS *****\n\n"

        "The student-run undergraduate philosophy journal Aperto Animo welcomes submissions to its {6} issue."
        " This is an excellent opportunity for undergrads to get published and share their ideas."
        " Submissions will be accepted on a rolling basis until {7}.\n\n"

        "Though expressly a philosophy journal, Aperto Animo also publishes academic writing on theology, political science, and multidisciplinary subjects." 
        " A strict focus on philosophy is not required.\n\n"
        
        "Consult the submission form linked below for more information.\n\n"
        
        "{8}\n\n"
        
        "Previous editions of Aperto Animo:\n\n"
        
        "https://philosophy.unc.edu/undergraduate/aperto-animo/\n\n"        

        "***** SAMPLE MESSAGE ENDS *****\n\n"
        
        "I hope that this opportunity will interest students at {2}."
        " Don't hesitate to reach out with questions. \n\n"

        "Best regards,\n\n"
        
        "{0}\n"
        "Editor-in-Chief, Aperto Animo\n"
        "{1}"
                  
        # {0} your name, {1} your email
        # {2} their university, {3} their department, {4} their title, {5} their email
        # {6} publication year, {7} submission deadline, {8} form link
        .format(your_name, your_email, university, department, title, email, publication_year, deadline, link)).encode('utf-8')

    # try sending the email
    try:
        # if the email can be sent, print a confirmation message
        server.sendmail(your_email, [email], full_email)
        print('Email to {} successfully sent!\n\n'.format(email))
        # if the email can't be sent, print an error message
    except Exception as e:
        print('Email to {} could not be sent :( because {}\n\n'.format(email, str(e)))

# exit your Gmail account
server.close()
```
